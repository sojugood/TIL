## B. 데이터 전처리 - 2021년 이후의 종가 데이터 출력하기
  - 2021년 이후 데이터만 필터링하기 위해 df에 조건을 달아줌

```python
# 패키지 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
csv_path = "data/NFLX.csv"

# CSV 파일 읽어오기 (마지막 두 개 열 제외)
df = pd.read_csv(csv_path, usecols=range(0, 5))

# 날짜 데이터 변환
df["Date"] = pd.to_datetime(df["Date"])

# 2021년 이후 데이터만 필터링
df = df[df["Date"] >= pd.to_datetime('2021-01-01')]

# 그래프 그리기 (가로, 세로 축에 표시될 데이터를 차례로 기입)
plt.plot(df['Date'], df['Close'])

# 그래프 제목 설정
plt.title('NFLX Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# y축 레이블 설정
plt.ylabel('Close Price')

# x축 설정(회전시키기)
plt.xticks(rotation=45)

# 그래프 표시
plt.show()
```

## D. 데이터 분석 - 2021년 이후 월 별 평균 종가 출력하기
  - 'Close' 데이터를 월별로 그룹화 하기 위해 'Date' 열을 인덱스로 설정하고, resample 함수를 이용하여 월별로 그룹화하고, mean 함수를 이용하여 각 그룹의 평균을 계산

```python
# 패키지 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
csv_path = "data/NFLX.csv"

# CSV 파일 읽어오기 (마지막 두 개 열 제외)
df = pd.read_csv(csv_path, usecols=range(0, 5))

# 날짜 데이터 변환
df["Date"] = pd.to_datetime(df["Date"])

# 2021년 이후 데이터만 필터링
df = df[df["Date"] >= pd.to_datetime('2021-01-01')]

# 인덱스를 날짜로 설정
df.set_index('Date', inplace=True)

# 월별로 그룹화하여 평균 종가 계산
monthly_df = df['Close'].resample('M').mean()

# 그래프 그리기
plt.plot(monthly_df.index, monthly_df.values)

# 그래프 제목 설정
plt.title('Monthly Average Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# y축 레이블 설정
plt.ylabel('Average Close Price')

# x축 설정(회전시키기)
plt.xticks(rotation=45)

# 그래프 표시
plt.show()
```

## E. 데이터 시각화 - 2022년 이후 월 별 최고, 최저, 종가 시각화하기

```python
# 패키지 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
csv_path = "data/NFLX.csv"

# CSV 파일 읽어오기 (마지막 두 개 열 제외)
df = pd.read_csv(csv_path, usecols=range(0, 5))

# 날짜 데이터 변환
df["Date"] = pd.to_datetime(df["Date"])

# 2022년 이후 데이터만 필터링
df = df[df["Date"] >= pd.to_datetime('2022-01-01')]

# 그래프 그리기
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.plot(df['Date'], df['Close'], label='Close')

# 그래프 제목 설정
plt.title('High, Low, and Close Prices since January 2022')

# x축 레이블 설정
plt.xlabel('Date')

# y축 레이블 설정
plt.ylabel('Price')

# x축 설정(회전시키기)
plt.xticks(rotation=45)

# 범례 표시
plt.legend()

# 그래프 표시
plt.show()
```
